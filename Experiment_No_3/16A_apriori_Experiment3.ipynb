```python
dataset = [
    ['French Fry', 'Burger', 'Nutmeg', 'Cheese', 'Cola', 'Noodles'],
    ['French Fry', 'Apple', 'Cheese', 'Cola'],
    ['French Fry', 'Unicorn', 'Corn', 'Cheese', 'Noodles'],
    ['Dill', 'Burger', 'Nutmeg', 'Cheese', 'Cola', 'Noodles'],
    ['French Fry', 'Apple', 'Cheese', 'Cola'],
    ['Dill', 'Burger', 'Nutmeg', 'Cheese', 'Cola', 'Noodles'],
    ['French Fry', 'Apple', 'Cheese', 'Cola'],
    ['French Fry', 'Unicorn', 'Corn', 'Cheese', 'Noodles'],
    ['Dill', 'Burger', 'Nutmeg', 'Cheese', 'Cola', 'Noodles'],
    ['French Fry', 'Apple', 'Cheese', 'Cola'],
    ['Corn', 'Burger', 'Burger', 'Cheese', 'Ice cream', 'Cola']
]

dataset
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori 
frequent_itemsets=apriori(df,min_support=0.6,use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
res=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.7)
res
```

```python
res1=res[['antecedents','consequents','support','confidence','lift']]
res1
```

```python
res2=res1[res1['confidence']>=1]
res2
```
